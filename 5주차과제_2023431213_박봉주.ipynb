```python
# filter() 함수: 자료형의 모든 항목을 순회하되 조건식이 참인 경우만 해당 값을 반환한다.
# 문제 15에서는 filter() 함수와 lambda 함수의 문법을 활용한다.

files = ['color', 'love.png', 'sky.jpg', 'cloud.xlsx', 'report.hwp']
x = list(filter(lambda each: (each[-3::] == 'png') or (each[-3::] == 'jpg'), files ))
print(x)
```
