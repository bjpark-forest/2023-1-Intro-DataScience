###문제4
아이리스 붓꽃 데이터를 이용하여 다음 선 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('04_data2.csv')
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']
```

```python
plt.figure(figsize=(15,4))
plt.plot(data.index, PetalLength, c='r', marker='o', label='Petal Length')
plt.plot(data.index, PetalWidth, c='g', marker='o', label='Petal Length')
plt.plot(data.index, SepalLength, c='b', marker='o', label='Petal Length')
plt.plot(data.index, SepalWidth, c='y', marker='o', label='Petal Length')


plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Line Graph')
plt.legend()
plt.grid()
plt.show()
```

###문제5
아이리스 붓꽃 데이터를 이용하여 다음 막대 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('04_data2.csv')
PetalLength = data['PetalLength'].mean()
PetalWidth = data['PetalWidth'].mean()
SepalLength = data['SepalLength'].mean()
SepalWidth = data['SepalWidth'].mean()
```

```python
plt.figure(figsize=(10,5))
plt.bar(["Petal Length", "Petal Width", "Sepal Length", "Sepal Width"], \
    [PetalLength,PetalWidth, SepalLength, SepalWidth ],\
    color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Iris')
plt.ylabel('Value')
plt.title('Iris Bar Graph')
plt.grid()
plt.show()
```

###문제6
아이리스 붓꽃 데이터를 이용하여 다음 산점도 그래프와 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('04_data2.csv')
SepalLength = data['SepalLength']
SepalWidth = data['SepalWidth']
```

```python
plt.figure(figsize=(10,4))

plt.scatter(SepalLength, SepalWidth)

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Sepal Scatter Graph')
plt.grid()
plt.show()
```

###문제7
아이리스 붓꽃 데이터를 이용하여 다음 히스토그램과 같이 출력되도록 빈칸에 알맞은 코드를 완성하시오.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
```

```python
virginica = data.loc[data.Species=='virginica',['Species','SepalLength']]

plt.figure(figsize=(10,6))
plt.hist(virginica.SepalLength, bins=7, label='Sepal Length')

plt.xlabel('Sepal')
plt.ylabel('Length')
plt.title('Iris Virginica Sepal Histogram')
plt.legend()
plt.grid()
plt.show()
```

###문제8
미세먼지 등급별 일수 데이터 또는 아이리스 붓꽃 데이터를 이용하여 히스토그램, 상자수염 그래프를 작성하시오.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('04_data2.csv')
```

```python
PetalLength = data['PetalLength']
PetalWidth = data['PetalWidth']

plt.figure(figsize=(8,5))
plt.boxplot([PetalLength, PetalWidth], labels=['Petal Length', 'Petal Width'])
plt.xlabel('Petal')
plt.ylabel('Value')
plt.title('Iris Petal Box Plot')
plt.show()
```
