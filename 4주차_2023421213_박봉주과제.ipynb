### 실습문제4

4.png

```python
import random
lotto = []
# 로또 번호를 생성하고, 생성된 로또 번호가 중복되는지 확인, 로또 번호 갯수가 모두 여섯 개인지 확인
while True:
  each = random.randint(1, 25)
  if each not in lotto:
    lotto.append(each)
  if len(lotto) == 6:
    break

print("<<생성된 로또 번호>>")
lotto.sort()
print(*lotto)
```

###문제11

11-2.png

```python
import random

answer = random.randint(1, 20)
#stop = True
while True:
  guess = int(input("숫자 입력(종료 0): "))
  if guess == 0:
    break
  elif guess == answer:
    print("정답!!")
    break
  elif guess > answer:
    print("더 작은 숫자 입력!")
  elif guess < answer:
    print("더 큰 숫자 입력!")
```

```python
import random

words = ["목소리", "기다림의", "소중해", "되고", "끝은"]
print("타자 게임 시작(엔터 입력)")

start = input()
quest_num = 0

while True:
    quest_num += 1       # 문제 번호 증가
    answer = random.choice(words)
    guide_text = f'문제 {quest_num}(종료 0): ' + answer 
    print(guide_text)
    repeat = input()

    if repeat == "0":    # 0이면 종료
      break
    elif repeat == answer:
      print("맞음!!")
    else:
      print("틀림! 다시!")
```
